### Encoding categorical data

```python
#converting the 'Cuisines ' column to lowercase for case insensitive matching
data1['Cuisines']=data1['Cuisines'].str.lower()
```

```python
data1
```

```python
#Determining the criteria for restaurant recommendations
user_cuisine_preference = 'north indian'
user_price_range_preference = 2.0 #assuming 2.0 correspods to moderate-priced restaurants
```

```python
#Implement Content based filtering using Decision Tree Regressor
#Encoding "Cuisines" using one-hot encoding
x=pd.get_dummies(data1['Cuisines'],drop_first=True)
y=data1['Aggregate rating']
                 
```

```python
x
```

```python
y
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

## Training the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
#Testing the recommendation system
user_cuisine_encoded = pd.get_dummies([user_cuisine_preference.lower()],prefix='',prefix_sep='').reindex(columns=x.columns,fill_value=0)
```

```python
#predicting the restaurant rating based on the user cuisine preference using the decision tree regressor model
predicted_rating=model.predict(user_cuisine_encoded)
```

```python
predicted_rating
```

```python
print(f"User Preference: Cuisine={user_cuisine_preference},Price Range={user_price_range_preference}")
print(f"Predicted Restaurant Rating: {predicted_rating[0]}")
```

## Evaluate the model

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
y_pred=model.predict(x_test)
```

```python
#mean_squared_error
print("MSE:",mean_squared_error(y_test,y_pred))
```

```python
#mean_absolute_error
print("MAE:",mean_absolute_error(y_test,y_pred))
```

```python
#R2_score
print("R2_score:",r2_score(y_test,y_pred))
```
